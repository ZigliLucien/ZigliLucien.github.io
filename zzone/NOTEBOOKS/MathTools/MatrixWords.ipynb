```python
s,t,u=symbols('s,t,u',commutative=False)
```

```python
Z=expand((s+t+u)**3)
Z
```

```python
A=Matrix(2,2,[0,1,1,1]);B=Matrix(2,2,[2,0,1,3]);C=Matrix(2,2,[0,1,1,0])
A,B,C
```

```python
pow=3
X=s*A+t*B+u*C
X2=(X**pow).expand()
X2
```

```python
L=list(itermonomials([s,t,u],pow)-itermonomials([s,t,u],pow-1))
#L=sorted(XL,key=monomial_key('lex',[s,t,u]))
L,L[0],len(L)
```

```python
B*C*A,Matrix(2,2,lambda i,j:X2[i,j].coeff(L[0])),Matrix(2,2,lambda i,j:X2[i,j].coeff(L[1])),C*C*B
```
