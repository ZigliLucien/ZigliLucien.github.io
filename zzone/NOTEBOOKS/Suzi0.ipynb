```python
def goPolys(f,z,N):
    C=zeros(N)
    for i in range(N):
        for j in range(N):
             C[i,j]=(diff(f,z[0],i,z[1],j)).subs([(z[0],0),(z[1],0)]).factor()
    return C   
```
