```python
from sympy import *
from IPython.display import *
#from sympy.polys.orderings import monomial_key
init_printing()

var('a:z');
```

```python
s,t,u=symbols('s,t,u',commutative=False)
```

```python
expand((s+t)**3)
```

```python
A=Matrix(2,2,[0,1,1,1]);B=Matrix(2,2,[2,0,1,3]);C=Matrix(2,2,[0,1,1,0])
A,B,C
```

```python
pow=3
X=s*A+t*B+u*C
X2=(X**pow).expand()
X2
```

```python
L=list(itermonomials([s,t,u],pow)-itermonomials([s,t,u],pow-1))
#L=sorted(IL,key=monomial_key('lex',[s,t,u]))
L,L[0]
```

```python
C**3,Matrix(2,2,lambda i,j:X2[i,j].coeff(L[0])),Matrix(2,2,lambda i,j:X2[i,j].coeff(L[1])),A*C*B
```

```python
M=[]
for i in range(len(L)):
    M.append([L[i],Matrix(2,2,lambda a,b:X2[a,b].coeff(L[i]))])
```
