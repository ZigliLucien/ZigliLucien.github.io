# Krawtchouk Polynomials

<pre>


</pre>

$$(1+v)^{N-j}(1-v)^j=\sum_{i=0}^N v^i\,\phi_{i}(j)$$

### where the $\phi_i$ are polynomials in $j$

Now we look at the polynomials as functions of $j$.


```python
def GenFun(q,dd):
    return series((1+v)**(q-j)*(1-v)**(j),v,0,dd+1)
```

```python
N=2
print('N='+str(N))
gg=GenFun(N,N)
ggg=[gg.coeff(v,i) for i in range(N+1)]
for i in range(N+1):
    display(ggg[i])
```

```python
display(Math('\,'))
display(Math("\phi_{"+str(0)+"}(j)="),ggg[0])
plot(ggg[0],(j,0,N),ylim=(0,2))
for i in range(1,len(ggg)):
        display(Math("\phi_{"+str(i)+"}(j)="),ggg[i])
        plot(ggg[i],(j,0,N))
```
